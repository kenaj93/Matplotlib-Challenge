## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress

```

```python
# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_info_df = pd.merge(study_results, mouse_metadata, how='left', on="Mouse ID")
# Display the data table for preview
mouse_info_df.head()
```

```python
# Checking the number of mice.
mice = mouse_info_df["Mouse ID"].nunique()
print(f"There is a total of {mice} mice.")
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
duplicates = mouse_info_df.loc[mouse_info_df.duplicated(subset=["Mouse ID", "Timepoint",]),"Mouse ID"].unique()

# Optional: Get all the data for the duplicate mouse ID. 
duplicate_id_df = mouse_info_df[mouse_info_df['Mouse ID'].isin(duplicates)==True]

duplicate_id_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_mouse_df = mouse_info_df[mouse_info_df['Mouse ID'].isin(duplicates)==False]

clean_mouse_df
```

```python
# Checking the number of mice in the clean DataFrame.
total_mice = clean_mouse_df["Mouse ID"].nunique()

print(f"There is a total of {total_mice} mice.")
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
#pull data
stats_df = clean_mouse_df.loc[:, ["Mouse ID", "Drug Regimen", "Tumor Volume (mm3)"]]

#calculate mean, median, variance, std, sem
# This method produces everything in a single groupby function
mean = stats_df.groupby(["Drug Regimen"]).mean()["Tumor Volume (mm3)"]
median = stats_df.groupby(["Drug Regimen"]).median()["Tumor Volume (mm3)"]
variance = stats_df.groupby(["Drug Regimen"]).var()["Tumor Volume (mm3)"]
deviation = stats_df.groupby(["Drug Regimen"]).std()["Tumor Volume (mm3)"]
SEM = stats_df.groupby(["Drug Regimen"]).sem()["Tumor Volume (mm3)"]



#create table
table_df = pd.DataFrame({"Mean":mean,"Median":median,"Standard Variance":variance,"Standard Deviation":deviation,"SEM":SEM})

table_df.round(2)
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
greatest_time = clean_mouse_df.groupby("Mouse ID").max().reset_index()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint

tumor_volume = greatest_time[["Mouse ID","Timepoint"]].merge(clean_mouse_df, on=["Mouse ID","Timepoint"], how='left')

tumor_volume.round(2)
```

```python
# Put treatments into a list for for loop (and later for plot labels)
capomulin= clean_mouse_df.loc[clean_mouse_df['Drug Regimen'] == 'Capomulin']['Tumor Volume (mm3)']
ramicane= clean_mouse_df.loc[clean_mouse_df['Drug Regimen'] == 'Ramicane']['Tumor Volume (mm3)']
infubinol=clean_mouse_df.loc[clean_mouse_df['Drug Regimen'] == 'Infubinol']['Tumor Volume (mm3)']
ceftamin= clean_mouse_df.loc[clean_mouse_df['Drug Regimen'] == 'Ceftamin']['Tumor Volume (mm3)']

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
# Locate the rows which contain mice on each drug and get the tumor volumes
# add subset 
# Determine outliers using upper and lower bounds

#Capomulin
quartile = capomulin.quantile([.25,.5,.75])
lower = quartile[.25]
upper = quartile[.75]
iqr= upper-lower
lower_out= lower-(1.5*iqr)
upper_out= upper+(1.5*iqr)    

#Ramicane
quartile_r = ramicane.quantile([.25,.5,.75])
lower_r = quartile_r[.25]
upper_r = quartile_r[.75]
iqr_r= upper_r-lower_r
lower_ra= lower_r-(1.5*iqr_r)
upper_ra= upper_r+(1.5*iqr_r)  

#infubinol
quartile_i = infubinol.quantile([.25,.5,.75])
lower_i = quartile_i[.25]
upper_i = quartile_i[.75]
iqr_i = upper_i-lower_i
lower_in= lower_i-(1.5*iqr_i)
upper_in= upper_i+(1.5*iqr_i)  


#Ceftamin
quartile_c = ceftamin.quantile([.25,.5,.75])
lower_c = quartile_c[.25]
upper_c = quartile_c[.75]
iqr_c= upper_c-lower_c
lower_ce= lower_c-(1.5*iqr_c)
upper_ce= upper_c+(1.5*iqr_c)  
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
box_it_out = dict(markerfacecolor='blue',markersize=5)

#Make Box Plot
plt.boxplot([capomulin,ramicane, infubinol, ceftamin], labels=["capomulin","ramicane","infubinol","ceftamin"], flierprops= box_it_out)

#Formatting
plt.title("Final Tumor Volumes")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

## Correlation and Regression
